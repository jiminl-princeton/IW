```python
# Sources:
# Mueller, A. (2012). word_cloud. GitHub. https://github.com/amueller/word_cloud
# Mueller, A. (2013). simple.py. GitHub. https://github.com/amueller/word_cloud/blob/master/examples/simple.py

# get data directory (using getcwd() is needed to support running example in generated IPython notebook)

# Read the whole text.
text = open("words_female.txt").read()
lines = text.split("\n")
d = {}
for line in lines:
  if line == "":
    continue
  word, freq = line.split(":")
  d[word] = int(freq)

# Generate a word cloud image
# Sources:
# NargesooTv. (2018). Create wordcloud from dictionary values (Do-me, Ed.). Stack Overflow. https://stackoverflow.com/questions/43145199/create-wordcloud-from-dictionary-values
wordcloud = WordCloud(background_color="white", max_words=100).generate_from_frequencies(d)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# # lower max_font_size
# wordcloud = WordCloud(max_font_size=40).generate(text)
# plt.figure()
# plt.imshow(wordcloud, interpolation="bilinear")
# plt.axis("off")
# Sources:
# Eytan, O. (2016). Matplotlib - How to plot a high resolution graph? (Martin, Ed.). Stack Overflow. https://stackoverflow.com/questions/39870642/matplotlib-how-to-plot-a-high-resolution-graph
plt.savefig('female.png', dpi=2500)

# render a single page (in this case: the first one)
plt.show()
```
