```python
import little_mallet_wrapper

# 15 topics
path_to_topic_keys_female = f"results/female/mallet.topic_keys.{str(15)}"
topics_female = little_mallet_wrapper.load_topic_keys(path_to_topic_keys_female)
words_female = []
for topic in topics_female:
  words_female += topic
words_female = set(words_female)

path_to_topic_keys_male = f"results/male/mallet.topic_keys.{str(15)}"
topics_male = little_mallet_wrapper.load_topic_keys(path_to_topic_keys_male)
words_male = []
for topic in topics_male:
  words_male += topic
words_male = set(words_male)

# Sources:
# Holtz, Y. (2018). Venn Diagram. The Python Graph Gallery. https://www.python-graph-gallery.com/venn-diagram/

venn = venn2([words_male, words_female], ('Male Corpus', 'Female Corpus'))
plt.show()
```
