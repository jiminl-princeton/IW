```python
text = open("authors_female.txt").read()
lines = text.split("\n")
authors_female = []
num_works_female = []
for line in lines[:25]:
  if line == "":
    continue
  author, num_works = line.split(":")
  if author[:3] == "Ram":
    author = "RamMariaLouiseOuidapseud"
  authors_female.append(author)
  num_works_female.append(int(num_works))
```

```python
# Sources:
# pandas. (2023). pandas.DataFrame.plot.barh. https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.barh.html
# Kassies, R. (2013). Setting Yaxis in Matplotlib using Pandas (gpr, Ed.). Stack Overflow. https://stackoverflow.com/questions/17787366/setting-yaxis-in-matplotlib-using-pandas

df = pd.DataFrame({'Author': authors_female, 'Number of Works': num_works_female})
ax = df.plot.barh(x='Author', y='Number of Works', rot=0)
ax.set_xticks(np.arange(0,15,1))
```
